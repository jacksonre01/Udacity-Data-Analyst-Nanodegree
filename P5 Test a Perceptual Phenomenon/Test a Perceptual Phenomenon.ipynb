(2) What is an appropriate set of hypotheses for this task? What kind of statistical test do you expect to perform? Justify your choices.

>Let **congruent-condition** equal the reading response time to the task in which the color word name and font color match  
Let **incongruent-condition** equal the reading response time to the task in which the color word name and font color do not match 

>**H0 - Null Hypothesis:** ( **μ**congruent-condition = **μ**incongruent-condition ) The population mean reading response time for the congruent (congruent-condition) words is _equal_ to the population mean reading response time for the incongruent (incongruent-condition) words.

>**HA - Alternative Hypothesis:** ( **μ**congruent-condition ≠ **μ**incongruent-condition ) The population mean reading response time for the congruent (congruent-condition) words is _significantly different_ from the population mean reading response time for the incongruent (incongruent-condition) words.

>$$H_0: μ_{congruent-condition} = μ_{incongruent-condition}$$

>$$H_A: μ_{congruent-condition} \neq μ_{incongruent-condition} $$

>**We have interval/continuous data, we are investigating the hypothesis of a difference in means, we have dependent samples, between two  paired (matched) measures, and our data based on the charts below appears to be  normally distrubuted therefore the appropriate test to use is the paired t-test.  Additionally, z-tests are best used for larger samples where n>30.  Our null hypothesis is that there is no difference between the mean response times for the congruent condition and the incongruent condition.  The image below represents the decision flow for chosing the t-test.**

(4) Provide one or two visualizations that show the distribution of the sample data. Write one or two sentences noting what you observe about the plot or plots.

```python
sns.set_style("darkgrid")
ax=sns.distplot(df['Congruent'])
plt.xlim(0,45)
plt.title('Reading Response Time vs Congruent Condition')
ax.set_xlabel('Reading Response Time')
plt.show()  
```

```python
sns.set_style("darkgrid")
ax=sns.boxplot(x="Incongruent", data=df,orient='h')
plt.xlim(0,45)
plt.title('Reading Response Time vs Incongruent Condition')
ax.set_xlabel('Reading Response Time')
plt.show()   
```

(5) Now, perform the statistical test and report the results. What is the confidence level and your critical statistic value? Do you reject the null hypothesis or fail to reject it? Come to a conclusion in terms of the experiment task. Did the results match up with your expectations?

```python

# Perform the statistical test here
scipy.stats.ttest_rel(df['Congruent'], df['Incongruent'])
```

>**Confidence level + alpha = 1. If we select alpha to equal .05, then we have 95% confidence level if our pvalue is less than alpha. If we allow alpha to be .01 then we have a 99% confidence level if our pvalue is less than alpha.  Our t-statistic is -8.021 and our pvalue is extremely small (small pvalues are good) and less than .01 which indicates that the results did not occur by chance and that we have a better than 99% confidence level.  Therefore we would reject our null hypothesis in favor of the alternate hypothesis that there is a statistically significant difference in the response rates.**
